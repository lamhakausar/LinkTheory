# Section 1A: The Triangle vs. Square vs. Chain – A Structural Comparison
 Core Claim: A triangle motif resists collapse better than squares and linear chains, because of its closed-loop tension and rigidity. This makes it the brain’s most stable building block.

## Triangle vs Square vs Chain: Structural Simulation

This section compares how different network shapes handle stress:
- The **triangle** is a closed, stable loop. Stress is redistributed evenly.
- The **square** is also closed but has more flexibility and is structurally weaker.
- The **chain** is open-ended; once one node fails, it can’t compensate from both sides.

Each node starts with a strength of 1.0. We stress one node and watch how the others are affected.
This helps us mathematically validate why triangle motifs are the most resilient microcircuits in Link Theory.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Create 3 small graphs
triangle = nx.Graph()
triangle.add_edges_from([(0, 1), (1, 2), (2, 0)])

square = nx.Graph()
square.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0)])

chain = nx.Graph()
chain.add_edges_from([(0, 1), (1, 2)])

motifs = {'Triangle': triangle, 'Square': square, 'Chain': chain}

# Simulate stress: reduce signal strength in node 0
initial_strength = {0: 1.0, 1: 1.0, 2: 1.0, 3: 1.0}

for name, G in motifs.items():
    strength = initial_strength.copy()
    strength[0] *= 0.4  # Stress node 0 (signal drop)

    # Neighbor nodes reduce based on number of connections
    for nbr in G.neighbors(0):
        strength[nbr] *= 0.7

    # Plot
    plt.figure(figsize=(4,4))
    pos = nx.spring_layout(G, seed=42)
    colors = [strength[n] if n in strength else 1.0 for n in G.nodes()]
    sizes = [strength[n]*1000 if n in strength else 1000 for n in G.nodes()]

    nx.draw(G, pos, with_labels=True, node_color=colors, node_size=sizes, cmap='coolwarm')
    plt.title(f"{name} Motif Stress Simulation")
    plt.show()
```

# Section 1B: Triangle Stability with Angles and Trigonometry
Core Claim:
Triangles are mechanically stable because their internal angles lock their shape — unlike squares or chains that can collapse without external support.

## Triangle Angle Sum and Structural Integrity

A triangle always has internal angles that sum to exactly 180°.  
This property creates a fixed structure — changing one side affects all angles.

In contrast, a square has angles that sum to 360°, and its sides can fold inward under force, making it less stable.

This is why triangle-based structures (like in geodesic domes) are used in architecture and nature — they redistribute force without collapse.


```python
import matplotlib.pyplot as plt
import numpy as np

# Coordinates of original triangle
triangle = np.array([[0, 0], [2, 0], [1, 2]])

# Simulate force on one vertex (slightly shift point C)
deformed_triangle = triangle.copy()
deformed_triangle[2] = [1.1, 2.1]  # applying force

# Coordinates of square
square = np.array([[0, 0], [2, 0], [2, 2], [0, 2]])

# Simulate force on one corner of the square (shift one point)
deformed_square = square.copy()
deformed_square[2] = [2.5, 1.5]  # applying force

# Plotting both shapes
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Triangle plot
axs[0].plot(*np.append(triangle, [triangle[0]], axis=0).T, label="Original", color='green')
axs[0].plot(*np.append(deformed_triangle, [deformed_triangle[0]], axis=0).T, linestyle='--', label="After Force", color='red')
axs[0].set_title("Triangle Before & After Force")
axs[0].set_aspect('equal')
axs[0].legend()
axs[0].grid(True)

# Square plot
axs[1].plot(*np.append(square, [square[0]], axis=0).T, label="Original", color='blue')
axs[1].plot(*np.append(deformed_square, [deformed_square[0]], axis=0).T, linestyle='--', label="After Force", color='orange')
axs[1].set_title("Square Before & After Force")
axs[1].set_aspect('equal')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

This visualization compares the structural response of a triangle vs. a square when one point is displaced (simulating stress):

The triangle maintains its overall shape and structural integrity. Even after applying force, it remains rigid with only minor distortion.

The square, however, deforms more easily — one corner collapses the structure, making it lose its original form.

This simple simulation supports a key claim of Link Theory:
Triangle motifs provide superior mechanical stability in networks — including brain networks — compared to other configurations like chains or squares.

This analysis provides a basic graph-theoretical tool to identify and assess triangle motifs in brain networks — central to modeling resilience in Link Theory.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Define a larger graph with multiple triangle motifs
edges = [
    (0,1), (1,2), (2,0),       # Triangle A
    (2,3), (3,4), (4,2),       # Triangle B (overlapping)
    (3,5), (5,6), (6,3),       # Triangle C
    (6,7), (7,8), (8,6),       # Triangle D
    (8,9), (9,0), (0,8)        # Triangle E (closing loop)
]

G = nx.Graph()
G.add_edges_from(edges)

# Assign initial signal strength
strength = {node: 1.0 for node in G.nodes()}

# Choose a node to stress
target = 3

# Stress simulation for 5 cycles
for cycle in range(5):
    print(f"\nCycle {cycle + 1}")

    # Decrease target node strength
    strength[target] *= 0.7

    # Neighbors absorb load → minor stress
    for neighbor in G.neighbors(target):
        strength[neighbor] *= 0.9

    # Display current strengths
    for node, val in strength.items():
        print(f"Node {node}: {val:.2f}")

    # Visualize
    sizes = [strength[n]*1000 for n in G.nodes()]
    colors = ['red' if n == target else 'orange' if strength[n] < 0.6 else 'green' for n in G.nodes()]

    plt.figure(figsize=(6,6))
    nx.draw(G, with_labels=True, node_size=sizes, node_color=colors)
    plt.title(f"Stress Propagation - Cycle {cycle + 1}")
    plt.show()
```

# Simulation: Stress Propagation in a Triangle-Rich Network
In this simulation, we model a simplified brain network composed of multiple overlapping triangle motifs — structures central to Link Theory. Each node starts with full signal strength (1.0), and one node is subjected to repeated stress. At each cycle:

The stressed node's strength decreases sharply (multiplied by 0.7).

Its immediate neighbors absorb load and decay more gradually (multiplied by 0.9).

The resulting weakening is visualized using node size and color, showing how stress spreads.

This example demonstrates a key principle of Link Theory:
Triangle motifs act as stabilizers, but sustained pressure on a single node can cause neighboring nodes to degrade, especially if rerouting is insufficient or repeated over time. This reflects a core concept of stress-induced failure in neural circuits.

```python
# Degree centrality: How connected each node is
degree_centrality = nx.degree_centrality(G)

# Triangle count per node
triangle_counts = nx.triangles(G)

# Display both metrics
print("Node\tDegree Centrality\tTriangles")
for node in G.nodes():
    print(f"{node}\t{degree_centrality[node]:.2f}\t\t\t{triangle_counts[node]}")
```

Together, these metrics help identify vulnerable vs. resilient nodes and support Link Theory’s claim that triangle-dense regions are key to maintaining brain function under stress.

#Modeling Rerouting in Link Theory


In this section, we extend our simulation by introducing a **basic rerouting mechanism** — a core concept of Link Theory. If a node's strength falls below a functional threshold (0.5), its triangle partners attempt to reroute signal through alternate paths, temporarily boosting signal in healthier nodes.

- **Rerouting Trigger**: If a triangle node falls below strength 0.5
- **Boost Factor**: Healthy adjacent nodes (strength > 0.6) receive a signal boost of 1.1×
- **Why it matters**: This reflects the brain's capacity to redistribute load after local failure — the beginnings of adaptive compensation before collapse.

This simulation provides a simplified but effective illustration of stress adaptation across triangle-based motifs, supporting Link Theory’s premise of network-based resilience.


```python
# Rerouting setup: continue from existing graph `G` and strength dictionary `node_strength`

reroute_threshold = 0.5
boost_factor = 1.1  # Simulated rerouting "boost"
target_node = 2  # If not already defined

for cycle in range(5, 10):  # Additional 5 cycles with rerouting
    print(f"--- Cycle {cycle+1} (with rerouting) ---")

    # Decay target node
    node_strength[target_node] *= 0.7

    for neighbor in G.neighbors(target_node):
        node_strength[neighbor] *= 0.85

        # If the target node is weak, try rerouting through neighbors' neighbors
        if node_strength[target_node] < reroute_threshold:
            for reroute_candidate in G.neighbors(neighbor):
                if reroute_candidate != target_node and node_strength[reroute_candidate] > 0.6:
                    node_strength[reroute_candidate] *= boost_factor
                    print(f"↪ Rerouted via Node {reroute_candidate} (boosted to {node_strength[reroute_candidate]:.3f})")

    # Print updated strengths
    for n in G.nodes():
        print(f"Node {n}: Strength = {node_strength[n]:.3f}")

    # Visualization
    plt.figure(figsize=(5, 5))
    sizes = [node_strength[n] * 800 + 100 for n in G.nodes()]  # Corrected size computation
    colors = ['red' if node_strength[n] < 0.5 else 'green' for n in G.nodes()]
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=True, node_color=colors, node_size=sizes)
    plt.title(f"Rerouting Simulation - Cycle {cycle+1}")
    plt.show()
```

 # Stress Propagation and Triangle Motif Failure

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create a triangle-rich graph
G = nx.Graph()
edges = [
    (0, 1), (1, 2), (2, 0),  # Triangle 1
    (1, 3), (2, 3),          # Triangle 2: 1-2-3
    (3, 4), (4, 5), (5, 3),  # Triangle 3
    (2, 6), (6, 7), (7, 2),  # Triangle 4
]
G.add_edges_from(edges)

# Initialize node strengths
strength = {node: 1.0 for node in G.nodes()}

# Identify triangle motifs
triangles = [c for c in nx.enumerate_all_cliques(G) if len(c) == 3]

# Stress parameters
target = 1
threshold = 0.3
cycles = 6
collapse_counts = []

# Simulation loop
for cycle in range(1, cycles + 1):
    print(f"\n--- Cycle {cycle} ---")

    # Stress the target node
    strength[target] *= 0.5

    # Spread stress to neighbors
    for neighbor in G.neighbors(target):
        strength[neighbor] *= 0.8

    # Check for collapsed triangles
    collapsed = 0
    for tri in triangles:
        if any(strength[n] < threshold for n in tri):
            collapsed += 1

    print(f"Collapsed triangle motifs: {collapsed}")
    collapse_counts.append(collapsed)
```

```python
import matplotlib.pyplot as plt

# Plotting the number of collapsed triangle motifs per cycle
plt.figure(figsize=(8, 5))
plt.plot(range(1, cycles + 1), collapse_counts, marker='o', color='crimson', linewidth=2)

plt.title("Triangle Motif Collapse Over Stress Cycles", fontsize=14)
plt.xlabel("Cycle Number", fontsize=12)
plt.ylabel("Number of Collapsed Triangle Motifs", fontsize=12)
plt.grid(True)
plt.xticks(range(1, cycles + 1))
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# Original x and y values
x = np.array(range(1, cycles + 1))
y = np.array(collapse_counts)

# Create a smooth spline curve
x_smooth = np.linspace(x.min(), x.max(), 300)
spline = make_interp_spline(x, y, k=2)  # quadratic smoothing
y_smooth = spline(x_smooth)

# Plot the smoothed curve
plt.figure(figsize=(8, 5))
plt.plot(x_smooth, y_smooth, color='crimson', linewidth=2)
plt.title("Smoothed Triangle Motif Collapse Over Stress Cycles", fontsize=14)
plt.xlabel("Cycle Number", fontsize=12)
plt.ylabel("Number of Collapsed Triangle Motifs", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.step(range(1, cycles + 1), collapse_counts, where='mid', color='crimson', linewidth=2)
plt.fill_between(range(1, cycles + 1), collapse_counts, step='mid', alpha=0.2, color='crimson')
plt.title("Stepwise Triangle Motif Collapse Over Cycles", fontsize=14)
plt.xlabel("Cycle Number")
plt.ylabel("Collapsed Motifs")
plt.grid(True)
plt.tight_layout()
plt.show()
```


This simulation numerically demonstrates Link Theory's key idea: triangle motifs initially resist collapse due to structural stability, but once a node weakens past a critical point, the failure spreads. Each cycle represents rising stress, and the increase in collapsed motifs reflects cascading structural failure.

By simulating the degradation of neural motifs under repeated stress cycles, we observe that collapse behaves non-linearly. The smoothed version highlights the system’s increasing fragility over time. The stepwise version further supports the theory that collapse occurs in punctuated thresholds rather than gradual decay — mirroring findings in trauma research, resilience literature, and nonlinear systems theory.
